```python
for filename in os.listdir(os.getcwd()):
    if filename=='ntuple_1.root':
        dfgen=uproot.open(filename)['Floatingpoint8ThresholdRef2dRef3dGenmatchGenclustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_gen,flatten=True)
        df_1=uproot.open(filename)['Floatingpoint8ThresholdRef2dRef3dGenmatchGenclustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_T23,flatten=True)
        df_2=uproot.open(filename)['Floatingpoint7ThresholdDummyHistomaxth10GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_3=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxth0GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_3=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxth10GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_4=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxth20GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_5=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxdr25GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_6=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxdr50GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_7=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxdr100GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_8=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxth0GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_9=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxth10GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_10=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxth20GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_11=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxdr25GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_12=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxdr50GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_13=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxdr100GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
    else:
        dfgen_=uproot.open(filename)['Floatingpoint8ThresholdRef2dRef3dGenmatchGenclustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_gen,flatten=True)
        df_1_=uproot.open(filename)['Floatingpoint8ThresholdRef2dRef3dGenmatchGenclustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_T23,flatten=True)
        df_2_=uproot.open(filename)['Floatingpoint7ThresholdDummyHistomaxth10GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_3_=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxth0GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_3_=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxth10GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_4_=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxth20GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_5_=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxdr25GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_6_=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxdr50GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_7_=uproot.open(filename)['Floatingpoint8SupertriggercellDummyHistomaxdr100GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_8_=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxth0GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_9_=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxth10GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_10_=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxth20GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_11_=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxdr25GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_12_=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxdr50GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        df_13_=uproot.open(filename)['Floatingpoint8ThresholdDummyHistomaxdr100GenmatchClustersntuple;1/HGCalTriggerNtuple'].pandas.df(branches_cl3d,flatten=True)
        
        dfgen=pd.concat(dfgen,dfgen_)
        df_1=pd.concat(df_1,df_1_)
        df_2=pd.concat(df_2,df_2_)
        df_3=pd.concat(df_3,df_3_)
        df_4=pd.concat(df_4,df_4_)
        df_5=pd.concat(df_5,df_5_)
        df_6=pd.concat(df_6,df_6_)
        df_7=pd.concat(df_7,df_7_)
        df_8=pd.concat(df_8,df_8_)
        df_9=pd.concat(df_9,df_9_)
        df_10=pd.concat(df_10,df_10_)
        df_11=pd.concat(df_11,df_11_)
        df_12=pd.concat(df_12,df_12_)
        df_13=pd.concat(df_13,df_13_)

        
algo={}
algo[0]=df_1
algo[1]=df_2
algo[2]=df_3
algo[3]=df_4
algo[4]=df_5
algo[5]=df_6
algo[6]=df_7
algo[7]=df_8
algo[8]=df_9
algo[9]=df_9
algo[10]=df_10
algo[11]=df_11
algo[12]=df_12
algo[13]=df_13

algo_clean={}
algo_name={}
algo_name[0]='algo0'
algo_name[1]='algo1'
algo_name[2]='algo2'
algo_name[3]='algo3'
algo_name[4]='algo4'
algo_name[5]='algo5'
algo_name[6]='algo6'
algo_name[7]='algo7'
algo_name[8]='algo8'
algo_name[9]='algo9'
algo_name[10]='algo10'
algo_name[11]='algo11'
algo_name[12]='algo12'
algo_name[13]='algo13'

#clean gen from particles that are not the originals or didn't reach endcap
sel=gen['genpart_reachedEE']==2 
gen_clean=gen[sel]
sel=gen_clean['genpart_gen']!=-1
gen_clean=gen_clean[sel]

#split df_gen_clean in two by eta sign
sel1=gen_clean['genpart_exeta']<=0
sel2=gen_clean['genpart_exeta']>0
gen_neg=gen_clean[sel1]
gen_pos=gen_clean[sel2]

gen_pos.set_index('event', inplace=True)
gen_neg.set_index('event', inplace=True)
thr=0.05
def matching(event):
    return event.cl3d_pt==event.cl3d_pt.max()


n_rec={}
deltar={}
#preprocessing for every algo
for i in algo:
    #split clusters in two by eta sign
    sel1=algo[i]['cl3d_eta']<=0
    sel2=algo[i]['cl3d_eta']>0
    algo_neg=algo[i][sel1]
    algo_pos=algo[i][sel2]
    #set the indices
    algo_pos.set_index('event', inplace=True)
    algo_neg.set_index('event', inplace=True)
    #merging
    algo_pos_merged=gen_pos.join(algo_pos, how='left', rsuffix='_algo')
    algo_neg_merged=gen_neg.join(algo_neg, how='left', rsuffix='_algo')
    #calculate deta and phi
    algo_pos_merged['deta']=algo_pos_merged['cl3d_eta']-algo_pos_merged['genpart_exeta']
    algo_pos_merged['dphi']=np.abs(algo_pos_merged['cl3d_phi']-algo_pos_merged['genpart_exphi'])

    algo_neg_merged['deta']=algo_neg_merged['cl3d_eta']-algo_neg_merged['genpart_exeta']
    algo_neg_merged['dphi']=np.abs(algo_neg_merged['cl3d_phi']-algo_neg_merged['genpart_exphi'])
    
    #get dphi between -pi and pi
    sel=algo_pos_merged['dphi']>np.pi
    algo_pos_merged['dphi']-=sel*(2*np.pi)
    algo_pos_merged['deltar']=np.sqrt(algo_pos_merged['dphi']*algo_pos_merged['dphi']+algo_pos_merged['deta']*algo_pos_merged['deta'])
    deltar_pos=algo_pos_merged['deltar']
    
    #get dphi between -pi and pi
    sel=algo_neg_merged['dphi']>np.pi
    algo_neg_merged['dphi']-=sel*(2*np.pi)
    algo_neg_merged['deltar']=np.sqrt(algo_neg_merged['dphi']*algo_neg_merged['dphi']+algo_neg_merged['deta']*algo_neg_merged['deta'])
    deltar_neg=algo_neg_merged['deltar']
    
    #keep the unreconstructed values (NaN)
    sel=pd.isna(algo_pos_merged['deltar']) 
    unmatched_pos=algo_pos_merged[sel]
    sel=pd.isna(algo_neg_merged['deltar'])  
    unmatched_neg=algo_neg_merged[sel]
    unmatched_pos['matches']=False
    unmatched_neg['matches']=False
    
    
    #select deltar under thr
    sel=algo_pos_merged['deltar']<=thr
    algo_pos_merged=algo_pos_merged[sel]
    sel=algo_neg_merged['deltar']<=thr
    algo_neg_merged=algo_neg_merged[sel]
    
    #matching
    group=algo_pos_merged.groupby('event')
    n_rec_pos=group['cl3d_pt'].size()
    algo_pos_merged['best_match']=group.apply(matching).array
    group=algo_neg_merged.groupby('event')
    n_rec_neg=group['cl3d_pt'].size()
    algo_neg_merged['best_match']=group.apply(matching).array
    #keep only matched clusters 
    sel=algo_pos_merged['best_match']==True
    algo_pos_merged=algo_pos_merged[sel]
    
    sel=algo_neg_merged['best_match']==True
    algo_neg_merged=algo_neg_merged[sel]
    
    #remerge with NaN values
    algo_pos_merged=pd.concat([algo_pos_merged, unmatched_pos], sort=False).sort_values('event') 
    algo_neg_merged=pd.concat([algo_neg_merged, unmatched_neg], sort=False).sort_values('event')

    
    
    deltar[i]=deltar_pos.append(deltar_neg)
    n_rec[i]=n_rec_pos.append(n_rec_neg)
    algo_clean[i]=pd.concat([algo_neg_merged,algo_pos_merged], sort=False).sort_values('event')

    algo_clean[i]['matches']=algo_clean[i]['matches'].replace(np.nan, True)
    algo_clean[i].drop(columns=['best_match'], inplace=True)
        
        
```
