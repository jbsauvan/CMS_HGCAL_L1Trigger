```python
def rmseff(x, c=0.68):
    """Compute half-width of the shortest interval containing a fraction 'c' of items in a 1D array."""
    x_sorted = np.sort(x, kind="mergesort") 
    m = int(c * len(x)) + 1
    return np.min(x_sorted[m:] - x_sorted[:-m]) / 2.0
```

```python
efficiencies_vs_pt = {}
for name,df in algo_match_unmatch.items():
    df['genmatch_pass'] = df['best_match']==True
    efficiencies_vs_pt[name] = df.groupby('genpart_binpt').mean()
    eff = df.groupby('genpart_binpt').apply(lambda x : efficiency(x, 'genmatch_pass'))
    efficiencies_vs_pt[name]['Reco_efficiency'] = eff
```
